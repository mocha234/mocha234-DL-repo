# DetectFace with Haar Cascade

### Packages Needed

```python
import numpy as np
import matplotlib.pyplot as plt
from cv2 import cv2

plt.rcParams["figure.figsize"] = (9,7)
```

### Display Image

```python
img = cv2.imread("hyunbinsonyejin.jpg")
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) 

plt.subplot(1, 1, 1) 
plt.imshow(img_rgb) 
plt.show()
```

### Cascade Classifier

```python
face_cascade = cv2.CascadeClassifier("haarcascade_frontalface_default.xml") 
```

### Convert BGR into Grayscale

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) 
```

```python
### Detect with defined Cascade Classifier
```

```python
found = face_cascade.detectMultiScale(img_gray,  
                                   minSize =(200, 200)) 
```

```python
amount_found = len(found) 
  
if amount_found != 0:    
    for (x, y, width, height) in found:   
        cv2.rectangle(img_rgb, (x, y),  
                      (x + height, y + width),  
                      (0, 255, 0), 5) 

plt.subplot(1, 1, 1) 
plt.imshow(img_rgb) 
plt.axis("Off")
plt.show() 
```
